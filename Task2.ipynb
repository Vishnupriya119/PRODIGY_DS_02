```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv('test.csv')
g_df=pd.read_csv('gender_submission.csv')
df
```

```python
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# Impute missing values for 'Age' with the median
df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

```python
# Impute missing values for 'Embarked' with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df
```

```python
# Drop 'Cabin' column due to a large number of missing values
df.drop('Cabin', axis=1, inplace=True)

```

```python
# Confirm that missing values have been handled
print("Missing Values After Imputation:")
print(df.isnull().sum())
```

```python
# Visualize the distribution of 'Age'
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Explore the relationship between 'Pclass' and 'Sex'
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Sex', data=df, palette='pastel')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Age', data=df, palette='pastel')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import os

# Specify the path to the folder containing the CSV files
folder_path = r'C:\Users\Vishnu Priya M\OneDrive\Documents\jupyter_demo\titanic'

# List all files in the folder
files = os.listdir(folder_path)

# Initialize empty DataFrames for each CSV filegende
passengers_df = pd.DataFrame()
survived_df = pd.DataFrame()
features_df = pd.DataFrame()

# Read each CSV file into its respective DataFrame
for file in files:
    if 'train' in file.lower():
        passengers_df = pd.read_csv(os.path.join(folder_path, file))
    elif 'test' in file.lower():
        features_df = pd.read_csv(os.path.join(folder_path, file))
    elif 'gender_submission' in file.lower():
        survived_df = pd.read_csv(os.path.join(folder_path, file))

# Display the first few rows of each DataFrame
print("Passengers DataFrame:")
print(passengers_df.head())

print("\nSurvived DataFrame:")
print(survived_df.head())

print("\nFeatures DataFrame:")
print(features_df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you've already loaded the Titanic dataset into three DataFrames: passengers_df, survived_df, features_df

# Combine the 'Survived' information with the 'Passengers' DataFrame
titanic_df = pd.merge(passengers_df, survived_df, on='PassengerId', how='left')

# Data Cleaning
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
```

```python
# Explore the relationship between 'Age' and 'Survived'
plt.figure(figsize=(8, 6))
g = sns.FacetGrid(titanic_df, hue='Survived_x', palette='pastel', height=6)
g.map(sns.histplot, 'Age', bins=20, kde=True)
g.add_legend()
plt.title('Survival Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
print(titanic_df.columns)

```

```python
# Explore the relationship between 'Fare' and 'Survived'
plt.figure(figsize=(8, 6))
sns.histplot(x='Fare', hue='Survived_x', data=titanic_df, bins=20, kde=True, palette='pastel')
plt.title('Survival Distribution by Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the relationship between 'Parch' (Number of Parents/Children) and 'Survived'
plt.figure(figsize=(8, 6))
sns.countplot(x='Parch', hue='Survived_x', data=titanic_df, palette='pastel')
plt.title('Survival Count by Number of Parents/Children')
plt.xlabel('Number of Parents/Children')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the relationship between 'SibSp' (Number of Siblings/Spouses) and 'Survived'
plt.figure(figsize=(8, 6))
sns.countplot(x='SibSp', hue='Survived_x', data=titanic_df, palette='pastel')
plt.title('Survival Count by Number of Siblings/Spouses')
plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the relationship between 'Pclass' and 'Fare'
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=titanic_df, hue='Pclass', palette='pastel', dodge=True, width=0.8, linewidth=0.8)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.legend(title='Passenger Class', loc='upper right')
plt.show()
```

```python
# Explore the relationship between 'Embarked' and 'Survived'
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', hue='Survived_x', data=titanic_df, palette='pastel')
plt.title('Survival Count by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the relationship between 'Sex', 'Pclass', and 'Survived'
plt.figure(figsize=(10, 8))
sns.catplot(x='Sex', hue='Survived_x', col='Pclass', data=titanic_df, kind='count', palette='pastel')
plt.title('Survival Count by Gender and Passenger Class')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the relationship between 'Fare', 'Age', and 'Survived' with a scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Age', y='Fare', hue='Survived_x', data=titanic_df, palette='bright', alpha=0.7)
plt.title('Scatter Plot of Fare vs Age by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Explore the count of passengers by 'Sex' and 'Survived' with a stacked bar chart
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived_x', data=titanic_df, palette='pastel')
plt.title('Count of Passengers by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Explore the distribution of fares by 'Embarked' with a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Embarked', y='Fare', hue='Embarked', data=titanic_df, palette='pastel', split=True, inner='quartile', legend=False)
plt.title('Distribution of Fares by Embarked Port')
plt.xlabel('Embarked Port')
plt.ylabel('Fare')
plt.show()

```

```python

# Explore pairwise relationships between numerical variables with a pair plot
sns.pairplot(titanic_df, hue='Survived_x', palette='bright', markers=['o', 's'], diag_kind='kde')
plt.suptitle('Pairwise Relationship between Numerical Variables by Survival', y=1.02)
plt.show()
```

```python
# Explore the distribution of 'Age' by 'Pclass' using a boxplot
plt.figure(figsize=(10, 8))
sns.boxplot(x='Pclass', y='Age', data=titanic_df, hue='Pclass', palette='bright', dodge=True, width=0.8, linewidth=0.8, legend=False)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# Explore the survival rate by 'Pclass' and 'Sex' with a clustered bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Pclass', y='Survived_x', hue='Sex', data=titanic_df, palette='bright')
plt.title('Survival Rate by Passenger Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Explore the count of passengers by 'SibSp' and 'Parch' with a clustered bar chart
plt.figure(figsize=(12, 8))
sns.countplot(x='SibSp', hue='Parch', data=titanic_df, palette='bright')
plt.title('Count of Passengers by Siblings/Spouses and Parents/Children')
plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Count')
plt.legend(title='Number of Parents/Children', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
